## Cleaning Data

```python
df2 = df.copy()
```

```python
df2["text"].head()
```

```python
df2["text"] = df2["text"].str.replace("(@+\w+)", "")
df2["text"].head()
```

```python
#nltk.download('stopwords')
#nltk.download('wordnet')
```

```python
stop_words = stopwords.words('english')
```

```python
def cleaning(data):
    
    #1. Tokenize
    text_tokens = word_tokenize(data.replace("'", "").lower()) 
     # ayralar kaldryorum ki (can't gibi) olumsuzluk ieren kelimeler iinde ayra ierdii iin stopwords lerde silinmesin.
        
    #2. Remove Puncs
    tokens_without_punc = [w for w in text_tokens if w.isalpha()]  # noktalama iaretlerinden temizleme
    
    #3. Removing Stopwords
    tokens_without_sw = [t for t in tokens_without_punc if t not in stop_words]  # stopword lerden temizleme
    # tokenleri teker teker al stopwords ler iinde yoksa olduu gibi yaz, yani stopwords ise yazma.
    
    #4. lemma
    text_cleaned = [WordNetLemmatizer().lemmatize(t) for t in tokens_without_sw]
    
    #joining
    return " ".join(text_cleaned)
```

```python
df2["text"] = df2["text"].apply(cleaning)
df2["text"].head()
```

```python
" ".join(df2["text"]).split()
```

```python
from sklearn.metrics import plot_confusion_matrix,classification_report, f1_score, recall_score
```

```python
def eval(model, X_train, X_test):
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    print("Test_Set")
    print(classification_report(y_test,y_pred))
    print("Train_Set")
    print(classification_report(y_train,y_pred_train))
    plot_confusion_matrix(model,X_test, y_test)
```

## Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB # BernoulliNB for binary model

# MultinomialNB, ikiden fazla label var ise bunun kullanlmas tavsiye ediliyor.

# yine de multinomial ve BernoilliNB den ikisini de deneyip hangisi daha iyi sonu veriyorsa onunla devam etmek gerekiyor.
# (nceden denenmi ve MulitnomialNB daha iyi km.)
```

```python
nb = MultinomialNB()
nb.fit(X_train_count,y_train)
```

```python
print("NB MODEL")
eval(nb, X_train_count, X_test_count)  # skorlara bakmak iin yukarda tanmladm eval fonksiyonunu kullanyorum. 

# BAKACAKLARIMIZ
# ncelikle F1_Score_Negative (Precision ve Recall Score larn Harmonic Ortalamas)

# Recall_Score_Negative (Negatif Skoru bulmada modelimin performans)(Olumsuz yorumlar)
# Precision_Score_Negative

# 3 skorunda iyi olmasn istiyorum. Modeli seerken buna dikkat ediyorum.

# Burada 3 hususa bakarak modelimize karar verdik. 
# 1-modelimin negative label lar ayrmadaki genel performans
# 2-F1 score (Harmonik ortalama (Recall-Precision))
# 3-Recall Score

# bizim iin model iyi demek ayn zamanda yapt tahminlerin de doru olmas demek
# hem F1 hem de Recal score yksek olsun istiyorum
# modelimin performans nemli. ona da bakmalym
# SVM altrmada maliyetli, logres daha eski ve insight daha fazla olduundan LOGREG TF-IDF seildi


# skorlarm dengeli. dolaysyla overfitting olmadn anlyoruz.
```

**Precision Recall Curve**
*   Unbalanced data da Precision Recall Curve kullanilir.
*   Sigmoid fonksiyonunun tresholdunu 0'dan 1'e arttirirken, her deger icin Precision ve Recall degerin gorsellestirilmesi ile elde edilen grafiktir.
*   Data balance olsayd Roc AUC Curve kullanacaktk.

```python
from yellowbrick.classifier import PrecisionRecallCurve   

viz = PrecisionRecallCurve(MultinomialNB(),
                           classes=nb.classes_,   # label larn isimleri ile gzkmesi iin.. aksi halde isim yerine 0,1,2 diye gzkr 
                           per_class=True,  # btn class lar grafikte gzksn diye True yaptk. False yapsaydk Micro-average PR for all classes olacakt.
                           cmap="Set1")
viz.fit(X_train_count,y_train)
viz.score(X_test_count, y_test)  # test datas zerinden skorlar alyoruz
viz.show();
```

```python
viz.score_
```


*   NEGATIVE class tahmin etmede olduka baarl (score 0,93)
*   POSITIVE class tahmin etmede olduka nispeten baarl (score 0,70)
*   NEUTRAL class tahmin etmede ayn baary gsterememi (0,58)
*   List item


```python
for i in ['negative','positive','neutral']:
  print(f"{i}: {viz.score_ [i]}")
```

```python
# Bunlar modellerin skorlarnn karlatrlmasnda kullanacaz. 

y_pred = nb.predict(X_test_count)
nb_count_rec_neg = recall_score(y_test, y_pred, labels = ["negative"], average = None)
nb_count_f1_neg = f1_score(y_test, y_pred, labels = ["negative"], average = None)
nb_AP_neg = viz.score_["negative"]

# labels parametresine birden fazla label yazarsak average onlarn ortalama skorlarn dndryor. o zaman average in bir anlam oluyor.
# fakat labels='a tek bir label yazdmzda average in de bir anlam olmuyor. (keza min, max vs. nin de anlam olmuyor)
# fonksiyon burda benden mutlaka bir average parametresi istedii iin average almasam da onu None olarak belirliyorum.  
```

# Cross Validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score
custom_scorer = {'accuracy': make_scorer(accuracy_score),
                 'precision-neg': make_scorer(precision_score, average=None, labels=["negative"]),
                 'recall-neg': make_scorer(recall_score, average=None, labels = ["negative"]),
                 'f1-neg': make_scorer(f1_score, average=None, labels = ["negative"]),
                 
                 'precision-neutral': make_scorer(precision_score, average=None, labels=["neutral"]),
                 'recall-neutral': make_scorer(recall_score, average=None, labels = ["neutral"]),
                 'f1-neutral': make_scorer(f1_score, average=None, labels = ["neutral"]),
                 
                 'precision-pos': make_scorer(precision_score, average=None, labels=["positive"]),
                 'recall-pos': make_scorer(recall_score, average=None, labels = ["positive"]),
                 'f1-pos': make_scorer(f1_score, average=None, labels = ["positive"]),
                 }

for i, j  in custom_scorer.items():
    model = LogisticRegression(C=0.4, max_iter=1000)
    scores = cross_val_score(model, X_train_tf_idf, y_train, cv = 10, scoring = j, n_jobs = -1).mean()  # cross val. train de oluyordu.
    print(f" {i:20} score : {scores}\n")
```
